```python
import pandas as pd
import matplotlib.pyplot as plt

# 폰트 설정 (Times New Roman)
plt.rcParams['font.family'] = 'Times New Roman'

# 파일 경로
file_path = '../earthquake_data/x_nan_2.xlsx'

# 데이터 로드
data = pd.read_excel(file_path, sheet_name=None)

# 'i2' 시트와 'o2' 시트 데이터 추출
i2_data = data['i2']
o2_data = data['o2']

# 각 시트의 첫 번째 열
i2_first_column = i2_data.iloc[:, 0]
o2_first_column = o2_data.iloc[:, 0]

# i2와 o2 시트의 첫 번째 열을 각각 X, Y 값으로 간주하여 scatter plot으로 그리기
plt.figure(figsize=(10, 6))
plt.scatter(i2_first_column, o2_first_column, color='green', alpha=0.6)

# 제목 및 레이블 설정
plt.title('Local_Magnitude vs PGA', fontsize=20)
plt.xlabel('Local_Magnitude', fontsize=20)
plt.ylabel('PGA', fontsize=20)

# X, Y 축 눈금 글씨 크기 조정
plt.tick_params(axis='both', which='major', labelsize=15)

# 그래프 표시
plt.tight_layout()
plt.show()
```
